## Paso 4: Buscar todos los empleos en la página

````python
# Aquí buscamos TODOS los contenedores que tienen empleos
# NOTA: Necesitas cambiar 'Card' por la clase correcta que encontraste

contenedores = sopa.find_all('div', class_='Card')
# find_all() = busca TODOS los elementos
# 'div' = tipo de etiqueta HTML
# class_='Card' = que tengan la clase "Card"

print(f"Se encontraron: {len(contenedores)} empleos")
print(f"\nPrimeros 3 empleos:")
for i, contenedor in enumerate(contenedores[:3], 1):
    print(f"\n{i}. {contenedor.prettify()[:300]}...")
````

### Conceptos importantes:

- **find_all()** = Busca todos los elementos que cumplan la condición
- **find()** = Busca solo el primero
- **class_** = Nota que es `class_` con subrayado (porque `class` es una palabra especial en Python)

---

## Paso 5: Extraer información de CADA empleo

````python
# Crearemos una lista para guardar todos los datos
empleos_lista = []

# Recorremos CADA contenedor de empleo
for i, contenedor in enumerate(contenedores, 1):
    try:
        # Extraer TÍTULO
        # Buscar la etiqueta que tiene el título (puede ser h3, h2, span, etc)
        titulo_elemento = contenedor.find('h3')  # Intenta encontrar h3
        titulo = titulo_elemento.text.strip() if titulo_elemento else "No disponible"
        # .text = obtiene el texto
        # .strip() = elimina espacios al inicio y final
        
        # Extraer FECHA DE PUBLICACIÓN
        fecha_elemento = contenedor.find('span', class_='fecha')
        fecha = fecha_elemento.text.strip() if fecha_elemento else "No disponible"
        
        # Extraer ACTIVO HASTA
        activo_elemento = contenedor.find('span', class_='activo')
        activo = activo_elemento.text.strip() if activo_elemento else "No disponible"
        
        # Extraer POSTULANTES
        postulantes_elemento = contenedor.find('span', class_='postulantes')
        postulantes = postulantes_elemento.text.strip() if postulantes_elemento else "No disponible"
        
        # Guardar en la lista
        empleos_lista.append({
            'Título': titulo,
            'Fecha Publicación': fecha,
            'Activo Hasta': activo,
            'Postulantes': postulantes
        })
        
        # Mostrar en pantalla
        print(f"{i}. {titulo}")
        print(f"   Publicado: {fecha} | Activo hasta: {activo} | Postulantes: {postulantes}\n")
        
    except Exception as error:
        print(f"Error en empleo {i}: {error}\n")

print(f"\n✓ Se extrajeron {len(empleos_lista)} empleos correctamente")
````

### Explicación línea por línea:

```python
for i, contenedor in enumerate(contenedores, 1):
```
- Recorre cada contenedor (cada empleo)
- `i` = número del empleo (1, 2, 3...)
- `enumerate()` = agrega números automáticamente

```python
titulo_elemento = contenedor.find('h3')
```
- Busca la etiqueta `<h3>` dentro del contenedor

```python
titulo = titulo_elemento.text.strip() if titulo_elemento else "No disponible"
```
- Si encontró algo: extrae el texto y elimina espacios
- Si NO encontró: pone "No disponible"
- Esto es un **condicional en una línea**
